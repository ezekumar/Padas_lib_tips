```python
imdb=pd.read_csv('http://bit.ly/imdbratings')
```

```python
imdb.head()
```

```python
boolean =[]

for durLength in imdb.duration:
    if durLength >= 200 :
        boolean.append(True)
    else:
        boolean.append(False)
    
```

```python
boolean[1:5]
```

```python
islong =pd.Series(boolean)
```

```python
imdb[islong]
```
